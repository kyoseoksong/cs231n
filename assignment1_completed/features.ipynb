## Load data
Similar to previous exercises, we will load CIFAR-10 data from disk.

```python
from cs231n.features import color_histogram_hsv, hog_feature

def get_CIFAR10_data(num_training=49000, num_validation=1000, num_test=1000):
  # Load the raw CIFAR-10 data
  cifar10_dir = 'cs231n/datasets/cifar-10-batches-py'
  X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)
  
  # Subsample the data
  mask = range(num_training, num_training + num_validation)
  X_val = X_train[mask]
  y_val = y_train[mask]
  mask = range(num_training)
  X_train = X_train[mask]
  y_train = y_train[mask]
  mask = range(num_test)
  X_test = X_test[mask]
  y_test = y_test[mask]

  return X_train, y_train, X_val, y_val, X_test, y_test

X_train, y_train, X_val, y_val, X_test, y_test = get_CIFAR10_data()

print X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape, y_test.shape
```

## Extract Features
For each image we will compute a Histogram of Oriented
Gradients (HOG) as well as a color histogram using the hue channel in HSV
color space. We form our final feature vector for each image by concatenating
the HOG and color histogram feature vectors.

* 각각의 이미지에 HOG와 Color Histogram 기법을 적용하여 feature를 추출한다.
* 최종 feature vector = HOG feature vector + Color Histogram feature vector (concatenation)

Roughly speaking, HOG should capture the texture of the image while ignoring
color information, and the color histogram represents the color of the input
image while ignoring texture. As a result, we expect that using both together
ought to work better than using either alone. Verifying this assumption would
be a good thing to try for the bonus section.

* HOG는 이미지의 texture feature 를 추출한다
* Color Histogram은 color feature를 추출한다.
* 이 코드에서는 2개를 concatenate해서 사용하지만, 각각 1개만 적용해서 (더 낮은) accuracy를 구해 보는 것도 해보면 보너스 점수!

The `hog_feature` and `color_histogram_hsv` functions both operate on a single
image and return a feature vector for that image. The extract_features
function takes a set of images and a list of feature functions and evaluates
each feature function on each image, storing the results in a matrix where
each column is the concatenation of all feature vectors for a single image.

* 3개의 function으로 구성된다
    * hog_feature(im) - return orientation_histogram.ravel() : texture feature vector 리턴
    * color_histogram_hsv(im, nbin=10, xmin=0, xmax=255, normalized=True) - return imhist : color feature vector 리턴
    * extract_features(imgs, feature_fns, verbose=False) - return imgs_features : concatenated feature vector 리턴

```python
from cs231n.features import *

num_color_bins = 20 # Number of bins in the color histogram
# (Andy) this is exactly the same number with extracted feature number from color_histogram_hsv()
# (Andy) if we change this number to 20, extracted feature number will be 20, too.

feature_fns = [hog_feature, lambda img: color_histogram_hsv(img, nbin=num_color_bins)] # 2
# (Andy) Usually we can say like this: feature_fns = [hog_feature, color_histogram_hsv]
# (Andy) By using lambda fn, we control the argument value for nbin

X_train_feats = extract_features(X_train, feature_fns, verbose=True) # (49000, 154)
# X_train_feats = extract_features(X_train[0:1000,:,:,:], feature_fns, verbose=True) # (49000, 154)
X_val_feats = extract_features(X_val, feature_fns) # (1000, 154)
X_test_feats = extract_features(X_test, feature_fns) # (1000, 154)
print X_train_feats.shape

# Preprocessing: Subtract the mean feature
mean_feat = np.mean(X_train_feats, axis=0, keepdims=True)
X_train_feats -= mean_feat # (49000, 154)
X_val_feats -= mean_feat # (1000, 154)
X_test_feats -= mean_feat # (1000, 154)

# Preprocessing: Divide by standard deviation. This ensures that each feature
# has roughly the same scale.
std_feat = np.std(X_train_feats, axis=0, keepdims=True)
X_train_feats /= std_feat # (49000, 154)
X_val_feats /= std_feat # (1000, 154)
X_test_feats /= std_feat # (1000, 154)

# Preprocessing: Add a bias dimension
X_train_feats = np.hstack([X_train_feats, np.ones((X_train_feats.shape[0], 1))]) #  # (49000, 155)
X_val_feats = np.hstack([X_val_feats, np.ones((X_val_feats.shape[0], 1))]) # (1000, 155)
X_test_feats = np.hstack([X_test_feats, np.ones((X_test_feats.shape[0], 1))]) # (1000, 155)
```

## Train SVM on features
Using the multiclass SVM code developed earlier in the assignment, train SVMs on top of the features extracted above; this should achieve better results than training SVMs directly on top of raw pixels.

```python
# Use the validation set to tune the learning rate and regularization strength

from cs231n.classifiers.linear_classifier import LinearSVM

learning_rates = [1e-9, 1e-8, 1e-7, 1e-6]
regularization_strengths = [1e4, 1e5, 1e6, 1e7, 1e8]

results = {} # key 는 (learning_rate, reg), value는 (train_accuracy, valid_accuracy)
best_val = -1 
best_svm = None

pass
################################################################################
# TODO:                                                                        #
# Use the validation set to set the learning rate and regularization strength. #
# This should be identical to the validation that you did for the SVM; save    #
# the best trained classifer in best_svm. You might also want to play          #
# with different numbers of bins in the color histogram. If you are careful    #
# you should be able to get accuracy of near 0.44 on the validation set.       #
################################################################################
for learning_rate in learning_rates:
    for reg in regularization_strengths:
        print learning_rate, reg
        svm = LinearSVM()
        loss = svm.train(X_train_feats, y_train, learning_rate=learning_rate, reg=reg,
                      num_iters=2000, verbose=True)
        y_train_pred = svm.predict(X_train_feats)
        train_accuracy = (np.mean(y_train == y_train_pred))
        y_val_pred = svm.predict(X_val_feats)
        valid_accuracy = (np.mean(y_val == y_val_pred))
        results[(learning_rate, reg)] = (train_accuracy, valid_accuracy)
        if best_val == -1 or valid_accuracy > best_val:
            best_val = valid_accuracy
            best_svm = svm
################################################################################
#                              END OF YOUR CODE                                #
################################################################################

# Print out results.
for lr, reg in sorted(results):
    train_accuracy, val_accuracy = results[(lr, reg)]
    print 'lr %e reg %e train accuracy: %f val accuracy: %f' % (
                lr, reg, train_accuracy, val_accuracy)
    
print 'best validation accuracy achieved during cross-validation: %f' % best_val
```

```python
# Evaluate your trained SVM on the test set
y_test_pred = best_svm.predict(X_test_feats)
test_accuracy = np.mean(y_test == y_test_pred)
print test_accuracy
```
